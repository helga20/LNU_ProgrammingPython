```python
import pandas as pd
from ipywidgets import widgets
from IPython.display import display
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("Airplane_Crashes_and_Fatalities_Since_1908.csv")
df = pd.DataFrame(df)
df.tail(5)
```

```python
df = df[['Location', 'Fatalities']]
df
```

```python
df = df.fillna(0)
```

```python
df.dtypes
```

```python
df = df.astype({df.columns[-1]: int})
df
```

```python
for index, row in df.iterrows(): # Ð¿ÐµÑÐµÐ±Ð¸ÑÐ°ÑÐ¼Ð¾ ÑÑÐ´ÐºÐ¸
    location = str(df.loc[index, "Location"])
    country = location.split(', ')[-1]
    df.loc[index, 'Location'] = country
df
```

```python
data1 = df.groupby('Location')['Fatalities'].sum().reset_index(name='Fatalities')
data1 = data1.sort_values(by=['Fatalities'])[-10:]
data1
```

```python
data2 = df.groupby('Location')['Location'].count().reset_index(name='Numbers')
data2 = data2.sort_values(by=['Numbers'])[-10:]
data2
```

```python
def plotter(what, how_many):
    fig, ax = plt.subplots()
    if what == 'Sum':
        ax.bar(data1['Location'][-how_many:], data1['Fatalities'][-how_many:])
        ax.set_title("Total Fatalities")
    else:
        ax.bar(data2['Location'][-how_many:], data2['Numbers'][-how_many:])
        ax.set_title("Total Aircrashes")
    plt.xticks(rotation=90)
    plt.show()


w = widgets.interactive(plotter, what=['Sum', 'Count'], how_many=(2, 10))
w
```
